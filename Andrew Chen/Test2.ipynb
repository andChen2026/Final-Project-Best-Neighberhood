This is a pie chart for measurements from neighborhood that belong in the moderate category

```python
Moderate = air_quality[(air_quality["index_value"] > 50) & (air_quality["index_value"] <= 100)]["site"]
moderate_sites_count = []

for x in sites:
    count = 0
    for y in Moderate:
        if x is y:
            count += 1
    moderate_sites_count.append(count)

mod_pie_dict = {
    "Sites": sites,
    "Counts": moderate_sites_count
}
mod_pie_chart = pd.DataFrame(mod_pie_dict)
plt.pie(mod_pie_chart["Counts"], labels=mod_pie_chart["Sites"])
plt.title("Moderate")
plt.show()
```

This is a pie chart for measurements from neighborhoods that belong in the unhealthy for some category

```python
Unhealthy_for_some = air_quality[(air_quality["index_value"] > 100) & (air_quality["index_value"] <= 150)]["site"]

ufs_sites_count = []

for x in sites:
    count = 0
    for y in Unhealthy_for_some:
        if x is y:
            count += 1
    ufs_sites_count.append(count)

ufs_pie_dict = {
    "Sites": sites,
    "Counts": ufs_sites_count
}
ufs_pie_chart = pd.DataFrame(ufs_pie_dict)
plt.pie(ufs_pie_chart["Counts"], labels=ufs_pie_chart["Sites"])
plt.title("Unhealthy For Some")
plt.show()
```

```python
Unhealthy = air_quality[(air_quality["index_value"] > 150) & (air_quality["index_value"] <= 200)]["site"]

u_sites_count = []

for x in sites:
    count = 0
    for y in Unhealthy:
        if x is y:
            count += 1
    u_sites_count.append(count)

u_pie_dict = {
    "Sites": sites,
    "Counts": u_sites_count
}
u_pie_chart = pd.DataFrame(u_pie_dict)
plt.pie(u_pie_chart["Counts"], labels=u_pie_chart["Sites"])
plt.title("Unhealthy")
plt.show()
```
